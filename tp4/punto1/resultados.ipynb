```python
import matplotlib.pyplot as plt# graficos
import numpy as np
```

#### Analisis Wiki-Small

```python
#Limit_Docs: 10% = 604 , 30% = 1812, 50%=3021, 100%=6043

times = {}
times[10] = [585.99957036972, 115.007452011108]
times[30] = [590.8260161876678, 95.57603645324707]
times[50] = [589.1423375606537, 84.59824466705322]
times[100] = [591.6369867324829, 75.48220944404602]
```

```python
times_index = [times[key][0] for key in times]
times_merge = [times[key][1] for key in times]
```

```python
plt.plot(times.keys(), times_index)
plt.title("Tiempos Indexado vs Límite de Documentos - WIKI-SMALL")
plt.xlabel("Limite de documentos(%)")
plt.ylabel("Tiempo de indexación(seg)")
plt.grid(True)
plt.show()
```

```python
plt.plot(times.keys(), times_merge)
plt.title("Tiempos Merge vs Límite de Documentos - WIKI-SMALL")
plt.xlabel("Limite de documentos(%)")
plt.ylabel("Tiempo de merge(seg)")
plt.grid(True)
plt.show()
```
